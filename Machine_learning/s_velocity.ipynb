```python
import numpy as np
import pandas as pd
```

```python
chars = ["array([[2]], dtype=uint8", "array", "\n", "[", "]", "(", ")"]

def string_to_ndarray(sd):
    for char in chars:
        sd = sd.replace(char, "")

    sd = sd.replace("   ", " ")
    sd = sd[:-2]

    sd_lst = sd.split(" ")
    sd_lst = [float(i) for i in sd_lst]

    sd_arr = np.asarray(sd_lst)
    sd_arr = np.split(sd_arr, len(sd_arr)/3)
    
    return np.asarray(sd_arr)
```

```python
df_data = pd.read_csv("../Data/preprocessed_conditiondata_vars.csv")
display(df_data)
```

```python
sd_norm1 = string_to_ndarray(df_data["SegmentData"][5])
sd_bril1 = string_to_ndarray(df_data["SegmentData"][622])
sd_norm2 = string_to_ndarray(df_data["SegmentData"][0])
sd_bril2 = string_to_ndarray(df_data["SegmentData"][616])
```
