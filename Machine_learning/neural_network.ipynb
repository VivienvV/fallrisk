```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier


pdata = pd.read_csv("../Data/preprocessed_alcohol_balance_perperson.csv")
display(pdata)

```

```python
# neural network arrays: 0 is no glasses, 1 is with glasses

xdf = pdata[['Conditie', 'mean_StepLength', 'Velocity', 'Balance_MeanVelocity']]
x = xdf.values
print(x)
#display(x)
```

```python
# Split the data into a training and validation set

def testround():
    def validation_split(data, ratio):
        split_point = int(ratio * np.size(data, 0))
        shuffled = np.random.shuffle(data)
        t_set = data[:split_point, :]
        val_set = data[split_point:, :]
        return (t_set, val_set)

    def valid_part(val_labels, predict_set):
        correct = 0
        dataset_size = predict_set.size
        for i in range(dataset_size):
            if predict_set[i] == val_labels[i]:
                correct += 1
        portion = float(correct)/float(dataset_size)
        return portion

    xnew = validation_split(x, 0.7)

    x_training = xnew[0]
    x_val = xnew[1]

    x_valtest = x_val[:, 1:]
    x_val_labels = x_val[:,0]

    y_training = x_training[:, 0]
    x_training = x_training[:, 1:]

    clf = MLPClassifier(solver='lbfgs')
    clf.fit(x_training, y_training)

    prediction_set = clf.predict(x_valtest)

  #  print(x_val_labels)
   # print(prediction_set)
    #print(valid_part(x_val_labels, prediction_set))
    
    return valid_part(x_val_labels, prediction_set)

print(testround())
```

```python
repeat = 50
testrounds = []

for i in range(repeat):
    Round = testround()
    testrounds.append(Round)
    
average = sum(testrounds) / float(len(testrounds))
print(average)
```
