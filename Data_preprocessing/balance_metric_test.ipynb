```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd

matdata = loadmat("../Data/SpineMidData_21032018.mat")
spineMidData=matdata['SpineMidData']
```

```python
csvdata = csv.reader(open('../Data/preprocessed_alcoholdata_sl_std.csv', 'r'), delimiter=',')

usableID = []

for row in csvdata:
    #for i in range(len(row)):
     #   if i is not 6:
      #      print(i, ':')
       #     print(row[i])
    nmbrsL = row[1]
    nmbrsL = nmbrsL[:-1]
    usableID.append(nmbrsL)

usableID.pop(0)
'''
for i in range(len(usableID)):
    print(i, ':')
    print(usableID[i])
'''
```

```python
glassesStart = 616

BMbegin = 13
BMend = 20
usableList = []

'''
for i in range(spineMidData[0, :].size):
    print(i,':')
    print(spineMidData[0, i])

print(spineMidData[0, 13:19])

'''


headers = np.hstack((spineMidData[0,0], spineMidData[0, BMbegin:BMend]))
headerlist = []
for i in range(headers.size):
    if i is 0:
        headerlist.append(headers[i])
    else:
        headerlist.append(headers[i][0])


for i in range(spineMidData[:,0].size):
    ID = spineMidData[i,0]
    if ID[0] in usableID:
        row = np.hstack((spineMidData[i,0], spineMidData[i, BMbegin:BMend]))
        usableList.append(row)

usableData = np.array(usableList)
#print(usableData)
```

```python
dataframe = pd.DataFrame.from_records(usableData, columns=headerlist)
display(dataframe)
```

```python
noglassData = usableData[:616, :]
glassData = usableData[616:, :]

noglassPanda = pd.DataFrame.from_records(noglassData, columns=headerlist)
glassPanda = pd.DataFrame.from_records(glassData, columns=headerlist)
display(noglassPanda)
```
