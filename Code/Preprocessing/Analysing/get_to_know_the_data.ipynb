```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data from matlab
data = loadmat("../Data/SpineMidData_21032018.mat")
```

```python
# get to know the keys
print('keys data:', data.keys())
spineMidData = data['SpineMidData']
print('keys spineMidData:\n', spineMidData[:1])
```

```python
# load data as panda 
pd_spineMidData = pd.read_csv("../Data/full_data.csv")
display(pd_spineMidData)
```

```python
# NUMBER OF UNIQUE VALUES
print('Number of unique values per column:')
for header in list(pd_spineMidData):
    print(header, pd_spineMidData[header].nunique())
```

```python
# RANGES
print('The data per column ranges between:')
for header in list(pd_spineMidData):
    print(header, pd_spineMidData[header].min(), '-', pd_spineMidData[header].max())
```

```python
# UNIQUE VALUES
print('Unique values per column:')
for header in list(pd_spineMidData):
    print(header, pd_spineMidData[header].unique())
```

```python
# BARPLOTS
for header in list(pd_spineMidData):
    perc = pd.DataFrame(pd_spineMidData[header].value_counts(normalize=True) * 100).sort_index()
    perc.plot.bar()
    plt.style.use('ggplot')
    plt.show()
```

```python
# PIECHARTS
headers = [
 "TrialNr",
 "NumberOfSegments",
 "SegmentNr",
 "Alcohol",
 "Height",
 "Weight",
 "Age'"]

for header in headers:
    perc = pd.DataFrame(pd_spineMidData[header].value_counts(normalize=True) * 100).sort_index()
    perc.plot.pie(subplots=True, autopct='%.2f', legend=False)
    plt.show()
```

```python
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

perc = pd.DataFrame(pd_spineMidData["TrialNr"].value_counts(normalize=True) * 100).sort_index()
explode = (0.1, 0, 0)
ax1.pie(perc, explode=explode)
ax1.axis('equal')
ax1.legend([1, 2, 3])
ax1.set_title("TrialNr")

perc = pd.DataFrame(pd_spineMidData["NumberOfSegments"].value_counts(normalize=True) * 100).sort_index()
explode = (0, 0, 0, 0.1, 0, 0)
ax2.pie(perc, explode=explode)
ax2.axis('equal')
ax2.legend([1, 2, 3, 4, 5, 6])
ax2.set_title("NumberOfSegments")

plt.show()
fig.savefig('preproprojection_np')    
```
